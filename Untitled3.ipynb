```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

# Generate mock data
random.seed(42)
np.random.seed(42)

users = 50
actions = ['login', 'view_resume_page', 'view_job_board', 'start_booking', 'complete_booking', 'logout']
devices = ['desktop', 'mobile', 'tablet']
start_time = datetime(2025, 5, 20, 8, 0)

data = []

for user_id in range(1, users + 1):
    session_start = start_time + timedelta(minutes=random.randint(0, 600))
    actions_per_user = random.sample(actions, random.randint(3, 6))
    device = random.choice(devices)
    for act in actions_per_user:
        timestamp = session_start + timedelta(minutes=random.randint(0, 30))
        data.append({
            'user_id': f'user_{user_id}',
            'timestamp': timestamp.strftime('%Y-%m-%d %H:%M:%S'),
            'action': act,
            'device': device
        })

df = pd.DataFrame(data)
df.to_csv("user_behavior_snapshot.csv", index=False)
df.head()
```

```python
df.to_csv('user_behavior_snapshot.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert timestamp to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['hour'] = df['timestamp'].dt.hour

# Plot: Most frequent actions
plt.figure(figsize=(8,4))
sns.countplot(data=df, x='action', order=df['action'].value_counts().index)
plt.title('Most Common User Actions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
